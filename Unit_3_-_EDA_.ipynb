```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
df_num = pd.read_csv("salary_numeric.csv")
df_num.head()
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import spearmanr
from scipy.stats import pearsonr
```

### What is the correlation between salary and education?

#### Your Hypothesis = A manager's salary increases with more college experience.

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df_num.education, df_num.salary)
```

There is a positive correlation between education and salary.

#### Are these results statistically significant? How do you know?

Yes because the P-value is zero.

### What is the correlation between salary and age?

#### Your Hypothesis = As age increases, salary increases moderately. 

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(df_num.salary, df_num.age)
```

There is a positive correlation between age and salary.

#### Are these results statistically significant? How do you know? 

Yes. The P-value is zero.

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = As industry experience increases, as does salary.

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
pearsonr(df_num.salary, df_num.industry_experience)
```

There is a positive correlation between salary and industry experience.

#### Are these results statistically significant? How do you know?

Yes because the P-values is zero.

#### Crosstab of average salary grouped by education and industry

```python
import numpy as np
```

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
crosstab = pd.crosstab(index=df_num['industry'], columns=df_num['education'], values=df_num['salary'], aggfunc=np.mean).round(0)
print(crosstab)
```
