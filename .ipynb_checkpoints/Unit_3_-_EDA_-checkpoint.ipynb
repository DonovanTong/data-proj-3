#### Fill in blanks with a brief description of each plot type:

  
- Histogram
> Histograms show the oscillation of data from one point to the next. It shows outliers a bit more precisely.
- Density Plot / KDE
> A density plot typically shows a smooth oscillation over time.
- Bar Chart
> Bar charts shows how data compares between different variables. It is used when a variable's position is more important than the value.
- Line Plot
> Used to show a generalized oscillation of data from one point to the next over time. 
- Univariate scatter plot
> Used when showing the change in a single variable.
- Violin Plot
> Used to show the frequency of an event occuring.
- Box Plot
> Used to show frequency of events as well as the median and range of data.
- Pie Chart
> Used to compare parts of a whole.
- Rug Plot
> Used to show the distribution of the data.

### Step 5 - Create Plots for Each Variable in Dataset

```python
# import seaborn
import seaborn as sns

# set standard plot type as seaborn plot
sns.set()

# import matplotlib
import matplotlib.pyplot as plt

# write command for plots to appear in jupyter cells automatically
%matplotlib inline
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### Plot for 'age' column

```python
# create plot for age column
df.age.plot(kind = "hist", figsize = (12,8), fontsize= 12, bins = 50)
plt.show()
```

##### Insights:

A histogram was much more effective in analyzing the age data than a bar or line chart. It helps to show the most common values.

### Plot for 'industry' column

```python
# create plot for industry column
sns.countplot(data=df, x='industry', palette='icefire')

plt.ylabel("Number of Managers in Industry")
plt.show()
```

##### Insights:

Seaborn is a super useful library. When comparing amounts in separate categories, a countplot is most useful.

### Plot for 'salary' column

```python
# create plot for salary column
df.salary.plot(kind="hist", figsize=(12,8), fontsize=12)
plt.show()
```

##### Insights:

Histogram plots are more effective for data associated with quantity at a certain rank.

### Plot for 'industry_experience' column

```python
df.industry_experience.plot(kind="hist", figsize=(12,8), fontsize=12)
plt.show()
```

##### Insights:

A histogram is a pretty universally useful plot method.

- mean
> simple average.
- median
> The most middle number.
- mode
> The number that occurs most often in the list
- central tendency
> the central position within a set of data (mean).
- distribution
> The occurence of different events throughout different variables.
- variance
> The spread between numbers in a data set
- standard deviation
> The amount which any given point in a data set deviates from the average.

### Use the Pandas Crosstab() function to compare categorical variables

#### crosstab of gender and education

```python
# use pd.crosstab to count number of managers at each education level, grouped by gender
pd.crosstab(index=df.gender, columns=df.education)
```

##### Insights:

The data is pretty evenly split by gender, but majority of managers graduated from college with some degree.

### Strip plot between age and education

```python
# create strip plot for age and education
sns.stripplot(data=df, x="education", y="age")
```

##### Insights:

Those with only high school experience only are hired and/or employed before the age of 30.

### Scatterplot for age and salary

```python
# create scatterplot for age and salary
sns.scatterplot(data=df, x="age", y="salary")
```

##### Insights:

The older the manager, likely, the more experience they have, which affects their salary heavily.

### Scatterplot for industry experience and salary

```python
df.industry_experience.describe()
```

```python
# create scatterplot for industry experience and salary
sns.scatterplot(data=df, x="industry_experience", y="salary")
```

##### Insights:
